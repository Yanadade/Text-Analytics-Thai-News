#Title Word Count Distribution


```python
from pythainlp import word_tokenize, sent_tokenize
from pythainlp import Tokenizer
from pythainlp.corpus import thai_stopwords
```

```python
import re
```

```python
def preprocess(text):
  clean_text = re.sub('[^ก-๙]',' ', text)
  clean_text = re.sub('[\.\/\~\-\_\=\!\?\"\“\”\(\)]+',' ',text)
  clean_text = re.sub('[\ๆ\ฯ]',' ',clean_text)
  clean_text = re.sub('[0-9]',' ',clean_text)
  token = word_tokenize(clean_text,engine='newmm')
  removestopword = ' '.join(i for i in token if not i in stopwords)
  return removestopword
```

```python
stopwords = list(thai_stopwords())
print(stopwords)
```

```python
df['clean_token'] = df['title'].apply(preprocess)
df['clean_token1'] = df['summary'].apply(preprocess)
df
```

```python
df = df.rename(columns = {'clean_token': 'title1','clean_token1': 'summary1' })
df
```
